```python
n = 100000
res = np.zeros(n)
lessThanSQRTThree = 0
for i in range(n):
    theta = random.uniform(0, 2*math.pi)
    x = math.cos(theta)
    y = math.sin(theta)
    chord = np.linalg.norm([x-1,y])
    res[i] = chord
    if chord < math.sqrt(3):
        lessThanSQRTThree += 1

print(np.mean(res))
print(lessThanSQRTThree/n)
sns.histplot(res)
```
